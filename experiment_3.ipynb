### Subtask:

**Reasoning**:
The first instruction is to define the `add_week_column` function as specified, which will convert a date column to datetime if necessary and calculate the start of the week.



````python
def add_week_column(df, date_column_name, new_column_name='week_start'):
    """
    Adds a 'week_start' column to the DataFrame, representing the start of each weekly window.

    Args:
        df (pd.DataFrame): The input DataFrame.
        date_column_name (str): The name of the column containing date information.
        new_column_name (str, optional): The name of the new column to store the week start. Defaults to 'week_start'.

    Returns:
        pd.DataFrame: The DataFrame with the new 'week_start' column.
    """
    # Ensure the date column is in datetime format
    df[date_column_name] = pd.to_datetime(df[date_column_name])

    # Calculate the start of the week (Monday as the start of the week)
    # .dt.to_period('W').dt.start_time sets the week to start on Monday
    # %w returns day of the week, where 0 is Sunday and 6 is Saturday.
    # Subtracting days to get to Monday. (date - timedelta(days=date.weekday()))
    df[new_column_name] = df[date_column_name].dt.to_period('W').dt.start_time

    print(f"'{new_column_name}' column added based on '{date_column_name}'.")
    return df

print("add_week_column function defined.")
````

## Weekly Aggregation: Aggregate Tweets and Events

### Subtask:
Implement functions to aggregate tweet data ('aggregate_weekly_tweets') and event data ('aggregate_weekly_events') into weekly summaries, including tweet volume, mean emotion probabilities, and event counts. Then, merge these into a single 'df_weekly' DataFrame using 'build_weekly_frame'.


**Reasoning**:
The first step is to define the `aggregate_weekly_tweets` function as per instruction 1. This function will group `df_clean` by `week_start`, calculate tweet volume, and count emotion labels for each week.



````python
def aggregate_weekly_tweets(df_clean):
    """
    Aggregates tweet data into weekly summaries, including tweet volume and emotion counts.

    Args:
        df_clean (pd.DataFrame): The DataFrame containing cleaned tweet data.

    Returns:
        pd.DataFrame: A DataFrame with weekly aggregated tweet data.
    """
    # Group by week_start and calculate tweet volume
    weekly_tweets = df_clean.groupby('week_start').size().reset_index(name='tweet_volume')

    # Calculate emotion counts for each week
    emotion_counts = pd.crosstab(df_clean['week_start'], df_clean['label'])

    # Merge tweet volume and emotion counts
    df_agg_tweets = pd.merge(weekly_tweets, emotion_counts, on='week_start', how='left')

    print("Weekly tweet data aggregated.")
    return df_agg_tweets

print("aggregate_weekly_tweets function defined.")
````

**Reasoning**:
Now that the `aggregate_weekly_tweets` function is defined, I will define the `aggregate_weekly_events` function as per instruction 2, which will group `df_acled` by `week_start` and count the events.



````python
def aggregate_weekly_events(df_acled):
    """
    Aggregates event data into weekly summaries, including event counts.

    Args:
        df_acled (pd.DataFrame): The DataFrame containing ACLED event data.

    Returns:
        pd.DataFrame: A DataFrame with weekly aggregated event data.
    """
    # Group by week_start and calculate event count
    df_agg_events = df_acled.groupby('week_start').size().reset_index(name='event_count')

    print("Weekly event data aggregated.")
    return df_agg_events

print("aggregate_weekly_events function defined.")
````

**Reasoning**:
Now that both `aggregate_weekly_tweets` and `aggregate_weekly_events` functions are defined, I will define the `build_weekly_frame` function as per instruction 3, which will merge the aggregated tweet and event dataframes.



````python
def build_weekly_frame(df_agg_tweets, df_agg_events):
    """
    Merges aggregated tweet and event data into a single weekly DataFrame.

    Args:
        df_agg_tweets (pd.DataFrame): DataFrame with weekly aggregated tweet data.
        df_agg_events (pd.DataFrame): DataFrame with weekly aggregated event data.

    Returns:
        pd.DataFrame: A merged DataFrame (df_weekly) with all weekly aggregated data.
    """
    # Merge the two aggregated DataFrames on the week_start column
    df_weekly = pd.merge(df_agg_tweets, df_agg_events, on='week_start', how='outer')

    # Fill any NaN values with 0 (for weeks with no tweets or no events)
    df_weekly = df_weekly.fillna(0)

    # Ensure 'week_start' is datetime for consistency if it became object after fillna
    df_weekly['week_start'] = pd.to_datetime(df_weekly['week_start'])

    print("Weekly tweet and event data merged into df_weekly.")
    return df_weekly

print("build_weekly_frame function defined.")
````

**Reasoning**:
Now that all three aggregation and merging functions (`aggregate_weekly_tweets`, `aggregate_weekly_events`, and `build_weekly_frame`) are defined, I will apply them in sequence as per instruction 4. This will create the `df_weekly` DataFrame. After creating `df_weekly`, I will display its first few rows and info as per instruction 5.



````python
import pandas as pd

# Apply aggregate_weekly_tweets to df_clean
df_agg_tweets = aggregate_weekly_tweets(df_clean)

# Apply aggregate_weekly_events to df_acled
df_agg_events = aggregate_weekly_events(df_acled)

# Apply build_weekly_frame to merge the aggregated data
df_weekly = build_weekly_frame(df_agg_tweets, df_agg_events)

print("\n--- df_weekly created ---")
print("\nFirst 5 rows of df_weekly:")
print(df_weekly.head())

print("\nInfo of df_weekly:")
df_weekly.info()
````

## Feature Engineering: Add Dynamic Features

### Subtask:
Implement and apply the `add_dynamic_features` function to calculate week-to-week changes, rolling means, and rolling standard deviations for key features in `df_weekly`.


**Reasoning**:
The first step is to define the `add_dynamic_features` function as per instruction 1, which will calculate week-to-week changes, rolling means, and rolling standard deviations for specified columns in `df_weekly`.



````python
def add_dynamic_features(df, rolling_window):
    """
    Calculates week-to-week changes, rolling means, and rolling standard deviations
    for key features in the DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame (df_weekly).
        rolling_window (int): The window size for rolling calculations.

    Returns:
        pd.DataFrame: The DataFrame with added dynamic features.
    """
    # Ensure 'week_start' is sorted for correct rolling calculations
    df = df.sort_values(by='week_start').reset_index(drop=True)

    # Columns for which dynamic features need to be calculated
    features_to_engineer = [
        'tweet_volume', 'anger', 'neutral', 'sarcasm', 'fear', 'event_count'
    ]

    for col in features_to_engineer:
        # Calculate week-to-week change
        df[f'{col}_change_weekly'] = df[col].diff()

        # Calculate rolling mean
        df[f'{col}_rolling_mean_{rolling_window}w'] = df[col].rolling(window=rolling_window).mean()

        # Calculate rolling standard deviation
        df[f'{col}_rolling_std_{rolling_window}w'] = df[col].rolling(window=rolling_window).std()

    # Fill NaN values resulting from diff and rolling operations with 0
    df = df.fillna(0)

    print(f"Dynamic features added with rolling window of {rolling_window}.")
    return df

print("add_dynamic_features function defined.")
````

````python
from sklearn.preprocessing import MinMaxScaler

def add_composite_signal(df):
    """
    Creates a normalized composite risk signal by combining selected features.

    Args:
        df (pd.DataFrame): The input DataFrame (df_weekly).

    Returns:
        pd.DataFrame: The DataFrame with the added 'composite_signal' column.
    """
    # Select key features (instruction 2)
    selected_features = ['anger', 'fear', 'event_count']

    # Create a new temporary column by summing these selected features (instruction 3)
    df['raw_composite'] = df[selected_features].sum(axis=1)

    # Normalize 'raw_composite' using Min-Max scaling (instruction 4)
    scaler = MinMaxScaler(feature_range=(0, 1))
    df['composite_signal'] = scaler.fit_transform(df[['raw_composite']])

    # Remove the temporary 'raw_composite' column (instruction 6)
    df = df.drop(columns=['raw_composite'])

    print("Composite signal added and normalized.")
    return df

print("add_composite_signal function defined.")
````

## Target Construction: Add Forecast Targets

### Subtask:
Implement and apply the `add_forecast_targets` function to create future-shifted target columns for both the composite signal and binary events for specified horizons (1, 2, and 3 weeks ahead).


**Reasoning**:
The first step is to define the `add_forecast_targets` function as per instruction 1. This function will first create a binary `event_occurred` column, then iterate through the specified horizons to create future-shifted target columns for the `composite_signal` and `event_occurred`. Finally, it will fill any NaN values resulting from the shifting operation with 0.

**Reasoning**:
The first step is to define the `add_forecast_targets` function as per instruction 1. This function will first create a binary `event_occurred` column, then iterate through the specified horizons to create future-shifted target columns for the `composite_signal` and `event_occurred`. Finally, it will fill any NaN values resulting from the shifting operation with 0.



````python
def add_forecast_targets(df, horizons):
    """
    Creates future-shifted target columns for composite signal and binary events.

    Args:
        df (pd.DataFrame): The input weekly DataFrame.
        horizons (list): A list of integer horizons (in weeks) for shifting.

    Returns:
        pd.DataFrame: The DataFrame with added future-shifted target columns.
    """
    # Ensure 'week_start' is sorted for correct shifting operations
    df = df.sort_values(by='week_start').reset_index(drop=True)

    # Create a binary column 'event_occurred' (Instruction 1.b)
    df['event_occurred'] = (df['event_count'] > 0).astype(int)

    # Create future-shifted target columns (Instruction 1.c)
    for h in horizons:
        # Target for composite signal
        df[f'composite_signal_target_h{h}'] = df['composite_signal'].shift(-h)
        # Target for binary events
        df[f'event_target_h{h}'] = df['event_occurred'].shift(-h)

    # Fill NaN values created by shifting with 0 (Instruction 1.d)
    df = df.fillna(0)

    print(f"Forecast target columns added for horizons: {horizons}.")
    return df

print("add_forecast_targets function defined.")
````

**Reasoning**:
Now that the `add_forecast_targets` function is defined, I will apply it to `df_weekly` with horizons `[1, 2, 3]` as per instruction 2. Then, I will display the first 5 rows and the info of the updated DataFrame to verify the new target columns as per instructions 3 and 4.



````python
horizons = [1, 2, 3]
df_weekly = add_forecast_targets(df_weekly, horizons)

print(f"\nFirst 5 rows of df_weekly after adding forecast targets for horizons {horizons}:")
print(df_weekly.head())

print(f"\nInfo of df_weekly after adding forecast targets for horizons {horizons}:")
df_weekly.info()
````

## Target Construction: Add Lag Features

### Subtask:
Implement and apply the `add_lag_features` function to generate lagged versions of core features, providing past context for forecasting models.


**Reasoning**:
The first step is to define the `add_lag_features` function as per instruction 1, which will generate lagged versions of core features in `df_weekly`.



````python
def add_lag_features(df, features_to_lag, lags):
    """
    Generates lagged versions of core features in the DataFrame.

    Args:
        df (pd.DataFrame): The input weekly DataFrame.
        features_to_lag (list): A list of column names to create lagged features for.
        lags (list): A list of integer lags (in weeks) to apply.

    Returns:
        pd.DataFrame: The DataFrame with added lagged features.
    """
    # Ensure 'week_start' is sorted for correct lagging operations (Instruction 2)
    df = df.sort_values(by='week_start').reset_index(drop=True)

    # Iterate through each feature and each lag (Instruction 3)
    for feature in features_to_lag:
        for lag in lags:
            # Create a new column named {feature}_lag_{lag} (Instruction 4)
            df[f'{feature}_lag_{lag}'] = df[feature].shift(lag)

    # Fill any NaN values created by shifting with 0 (Instruction 5)
    df = df.fillna(0)

    print(f"Lagged features added for {features_to_lag} with lags: {lags}.")
    return df

print("add_lag_features function defined.")
````

**Reasoning**:
Now that the `add_lag_features` function is defined, I will apply it to `df_weekly` with the specified `features_to_lag` and `lags`, then display the first few rows and the info of the updated DataFrame as per instructions 7, 8, 9, 10, and 11.



````python
features_to_lag = ['tweet_volume', 'anger', 'neutral', 'sarcasm', 'fear', 'event_count', 'composite_signal']
lags = [1, 2, 3, 4]

df_weekly = add_lag_features(df_weekly, features_to_lag, lags)

print(f"\nFirst 5 rows of df_weekly after adding lagged features (lags = {lags}):")
print(df_weekly.head())

print(f"\nInfo of df_weekly after adding lagged features (lags = {lags}):")
df_weekly.info()
````

````python
import numpy as np

def create_sequences_from_df(df, feature_cols, target_cols, sequence_length):
    """
    Prepares sequential data (features and targets) from a DataFrame for LSTM models.

    Args:
        df (pd.DataFrame): The input DataFrame.
        feature_cols (list): List of column names to be used as features.
        target_cols (list): List of column names to be used as targets.
        sequence_length (int): The length of each input sequence.

    Returns:
        tuple: A tuple containing (X, y) NumPy arrays, where X are the feature sequences
               and y are the corresponding target values.
    """
    X, y = [], []
    data = df[feature_cols].values
    targets = df[target_cols].values

    for i in range(len(df) - sequence_length):
        # Extract a sequence of `sequence_length` from `feature_cols` starting at `i`
        X.append(data[i:(i + sequence_length)])
        # Extract the corresponding target from `target_cols` at `i + sequence_length - 1`
        y.append(targets[i + sequence_length - 1])

    print(f"Created {len(X)} sequences of length {sequence_length}.")
    return np.array(X), np.array(y)

print("create_sequences_from_df function defined.")
````

## Event Warning: Convert Forecast to Alerts and Evaluate

### Subtask:
Implement functions `forecast_to_alerts` to convert continuous forecasts into binary early warning alerts, `evaluate_event_warnings` to assess alert quality (precision, recall, F1), and `compute_lead_time` to analyze warning lead times.


**Reasoning**:
The first step is to define the `forecast_to_alerts` function as per instruction 1, which will convert continuous forecasts into binary early warning alerts based on a given threshold.



````python
import numpy as np

def forecast_to_alerts(forecasts, threshold):
    """
    Converts continuous forecasts into binary early warning alerts.

    Args:
        forecasts (np.array): Array of continuous forecast values.
        threshold (float): The threshold above which an alert is triggered.

    Returns:
        np.array: Array of binary alerts (1 if forecast > threshold, 0 otherwise).
    """
    alerts = (forecasts > threshold).astype(int)
    print(f"Converted {len(forecasts)} forecasts to binary alerts using threshold {threshold}.")
    return alerts

print("forecast_to_alerts function defined.")
````

**Reasoning**:
Now that the `forecast_to_alerts` function is defined, I will define the `evaluate_event_warnings` function as per instruction 2. This function will calculate precision, recall, and F1-score for the early warning system using `sklearn.metrics`.



````python
from sklearn.metrics import precision_score, recall_score, f1_score

def evaluate_event_warnings(actual_events, predicted_alerts):
    """
    Assesses the quality of early warning alerts using precision, recall, and F1-score.

    Args:
        actual_events (np.array): Binary array indicating actual event occurrences.
        predicted_alerts (np.array): Binary array indicating predicted alerts.

    Returns:
        dict: A dictionary containing precision, recall, and F1-score.
    """
    # Ensure inputs are 1D arrays and of integer type
    actual_events = np.asarray(actual_events).flatten().astype(int)
    predicted_alerts = np.asarray(predicted_alerts).flatten().astype(int)

    # Handle cases where all actual_events or predicted_alerts are zero to avoid warnings
    if np.sum(actual_events) == 0 and np.sum(predicted_alerts) == 0:
        precision = 1.0 # No events, no alerts, perfect precision
        recall = 1.0    # No events, perfect recall
        f1 = 1.0
    elif np.sum(actual_events) == 0:
        precision = 0.0 # No actual events but alerts were raised
        recall = 0.0    # No actual events
        f1 = 0.0
    elif np.sum(predicted_alerts) == 0:
        precision = 0.0 # No alerts but there were actual events
        recall = 0.0    # No alerts
        f1 = 0.0
    else:
        precision = precision_score(actual_events, predicted_alerts, zero_division=0)
        recall = recall_score(actual_events, predicted_alerts, zero_division=0)
        f1 = f1_score(actual_events, predicted_alerts, zero_division=0)

    print("Early warning alerts evaluated.")
    return {'precision': precision, 'recall': recall, 'f1_score': f1}

print("evaluate_event_warnings function defined.")
````

**Reasoning**:
Now that the `forecast_to_alerts` and `evaluate_event_warnings` functions are defined, I will define the `compute_lead_time` function as per instruction 3. This function will calculate the average lead time of true positive alerts.



````python
def compute_lead_time(actual_events, predicted_alerts, horizon):
    """
    Calculates the average lead time of true positive alerts.

    Args:
        actual_events (np.array): Binary array indicating actual event occurrences.
        predicted_alerts (np.array): Binary array indicating predicted alerts.
        horizon (int): The forecast horizon in weeks.

    Returns:
        float: The average lead time in weeks for true positive alerts, or 0 if no true positives.
    """
    # Ensure inputs are 1D arrays and of integer type
    actual_events = np.asarray(actual_events).flatten().astype(int)
    predicted_alerts = np.asarray(predicted_alerts).flatten().astype(int)

    lead_times = []
    for i in range(len(actual_events)):
        # Check for a true positive: an alert was raised and an event actually occurred
        if predicted_alerts[i] == 1 and actual_events[i] == 1:
            # The lead time is the horizon since the alert was for 'horizon' weeks ahead
            lead_times.append(horizon)

    if lead_times:
        avg_lead_time = np.mean(lead_times)
    else:
        avg_lead_time = 0.0 # No true positives, so no lead time to report

    print(f"Average lead time computed: {avg_lead_time:.2f} weeks.")
    return avg_lead_time

print("compute_lead_time function defined.")
````

Implement `run_one_horizon` function to orchestrate the forecasting pipeline for a single horizon.


**Reasoning**:

````python
def run_one_horizon(train_df, val_df, test_df, feature_set, target_col, horizon, sequence_length, lstm_units, epochs, batch_size, arima_order, event_threshold):
    """
    Orchestrates the forecasting pipeline for a single horizon, including data preparation,
    model training, prediction generation, and event warning evaluation for multiple models.

    Args:
        train_df (pd.DataFrame): Training DataFrame.
        val_df (pd.DataFrame): Validation DataFrame.
        test_df (pd.DataFrame): Test DataFrame.
        feature_set (list): List of feature column names.
        target_col (str): The name of the target column (e.g., 'composite_signal_target_h1').
        horizon (int): The forecast horizon.
        sequence_length (int): The length of input sequences for LSTM.
        lstm_units (int): Number of LSTM units.
        epochs (int): Number of training epochs for LSTM.
        batch_size (int): Batch size for LSTM training.
        arima_order (tuple): ARIMA model order (p, d, q).
        event_threshold (float): Threshold for converting forecasts to binary alerts.

    Returns:
        dict: A dictionary containing predictions, alerts, evaluation metrics, and lead times
              for all models for the specified horizon.
    """
    print(f"\n--- Running pipeline for horizon {horizon}, target {target_col} ---")

    # 2. Prepare target series for ARIMA
    train_val_series = pd.concat([train_df['composite_signal'], val_df['composite_signal']])

    # 3. Extract true target values for regression evaluation
    # We need to make sure this aligns with the length of the predictions from LSTM (test_X length)
    # The LSTM predictions have a length of test_df_scaled - sequence_length + 1. The test_y is also this length.
    test_y_true_regression = test_df[target_col].values[sequence_length - 1:].flatten()

    # 4. Extract true binary event values for event warning evaluation
    event_target_col = f'event_target_h{horizon}'
    # Align actual events to the length of LSTM-generated data
    test_y_true_events = test_df[event_target_col].values[sequence_length - 1:].astype(int).flatten()

    results = {
        'horizon': horizon,
        'target_col': target_col,
        'true_regression': test_y_true_regression,
        'true_events': test_y_true_events,
        'models': {
            'persistence': {},
            'moving_average': {},
            'arima': {},
            'lstm': {},
            'hybrid': {}
        }
    }

    # 5. Run Persistence Model
    persistence_predictions = run_persistence(test_df, target_col, horizon)
    # Align persistence predictions to match the length of other models' predictions
    persistence_predictions_aligned = persistence_predictions.values[sequence_length - 1:]
    results['models']['persistence']['predictions'] = persistence_predictions_aligned
    results['models']['persistence']['alerts'] = forecast_to_alerts(persistence_predictions_aligned, event_threshold)
    results['models']['persistence']['metrics'] = evaluate_event_warnings(test_y_true_events, results['models']['persistence']['alerts'])
    results['models']['persistence']['lead_time'] = compute_lead_time(test_y_true_events, results['models']['persistence']['alerts'], horizon)


    # 6. Run Moving Average Model (using rolling_window = 4 as previously defined)
    moving_average_predictions = run_moving_average(test_df, target_col, horizon, rolling_window=4)
    # Align moving average predictions
    moving_average_predictions_aligned = moving_average_predictions.values[sequence_length - 1:]
    results['models']['moving_average']['predictions'] = moving_average_predictions_aligned
    results['models']['moving_average']['alerts'] = forecast_to_alerts(moving_average_predictions_aligned, event_threshold)
    results['models']['moving_average']['metrics'] = evaluate_event_warnings(test_y_true_events, results['models']['moving_average']['alerts'])
    results['models']['moving_average']['lead_time'] = compute_lead_time(test_y_true_events, results['models']['moving_average']['alerts'], horizon)

    # 7. Run ARIMA Model
    arima_predictions = run_arima(train_val_series, test_df['composite_signal'], horizon, arima_order)
    # Align ARIMA predictions
    arima_predictions_aligned = arima_predictions.values[sequence_length - 1:]
    results['models']['arima']['predictions'] = arima_predictions_aligned
    results['models']['arima']['alerts'] = forecast_to_alerts(arima_predictions_aligned, event_threshold)
    results['models']['arima']['metrics'] = evaluate_event_warnings(test_y_true_events, results['models']['arima']['alerts'])
    results['models']['arima']['lead_time'] = compute_lead_time(test_y_true_events, results['models']['arima']['alerts'], horizon)

    # 8. Scale features for LSTM
    train_df_scaled, val_df_scaled, test_df_scaled = scale_features(train_df, val_df, test_df, feature_set)

    # 9. Create LSTM sequences
    train_X, train_y = create_sequences_from_df(train_df_scaled, feature_set, [target_col], sequence_length)
    val_X, val_y = create_sequences_from_df(val_df_scaled, feature_set, [target_col], sequence_length)
    test_X, test_y_lstm = create_sequences_from_df(test_df_scaled, feature_set, [target_col], sequence_length)
    # Note: test_y_lstm is used for LSTM model's internal evaluation, but test_y_true_regression is for overall evaluation.

    # 10. Build LSTM model
    input_shape = (train_X.shape[1], train_X.shape[2])
    lstm_model = build_lstm_model(input_shape, lstm_units)

    # 11. Run LSTM Regression
    lstm_predictions = run_lstm_regression(train_X, train_y, val_X, val_y, test_X, test_y_lstm, lstm_model, epochs, batch_size, target_col)
    results['models']['lstm']['predictions'] = lstm_predictions
    results['models']['lstm']['alerts'] = forecast_to_alerts(lstm_predictions, event_threshold)
    results['models']['lstm']['metrics'] = evaluate_event_warnings(test_y_true_events, results['models']['lstm']['alerts'])
    results['models']['lstm']['lead_time'] = compute_lead_time(test_y_true_events, results['models']['lstm']['alerts'], horizon)

    # 12. Run Hybrid ARIMA-LSTM Model
    hybrid_predictions = run_hybrid_arima_lstm(pd.Series(arima_predictions_aligned), lstm_predictions, test_y_true_regression, target_col)
    results['models']['hybrid']['predictions'] = hybrid_predictions
    results['models']['hybrid']['alerts'] = forecast_to_alerts(hybrid_predictions, event_threshold)
    results['models']['hybrid']['metrics'] = evaluate_event_warnings(test_y_true_events, results['models']['hybrid']['alerts'])
    results['models']['hybrid']['lead_time'] = compute_lead_time(test_y_true_events, results['models']['hybrid']['alerts'], horizon)

    print(f"--- Pipeline finished for horizon {horizon}, target {target_col} ---")
    return results

print("run_one_horizon function defined.")
````

````python
import pandas as pd
import numpy as np

def audit_df(df, df_name, expected_shape=None):
    """
    Audits a pandas DataFrame, checking its shape, columns, nulls, and descriptive statistics.

    Args:
        df (pd.DataFrame): The DataFrame to audit.
        df_name (str): The name of the DataFrame for logging purposes.
        expected_shape (tuple, optional): A tuple (rows, cols) to compare the DataFrame's shape against.
    """
    print(f"--- Audit Report for DataFrame: {df_name} ---")
    print(f"Shape: {df.shape}")
    print("First 5 Rows:")
    print(df.head())
    print("\nDataFrame Info:")
    df.info()
    print(f"\nTotal Null Values: {df.isnull().sum().sum()}")

    if expected_shape is not None:
        match_status = "matches" if df.shape == expected_shape else "does NOT match"
        print(f"Expected Shape: {expected_shape}, Actual Shape {df.shape}. Shapes {match_status}.")

    print("\nDescriptive Statistics:")
    print(df.describe())
    print(f"--- End Audit Report for DataFrame: {df_name} ---\n")

def audit_array(arr, arr_name, expected_shape=None):
    """
    Audits a NumPy array, checking its shape, dtype, and first values.

    Args:
        arr (np.ndarray): The NumPy array to audit.
        arr_name (str): The name of the array for logging purposes.
        expected_shape (tuple, optional): A tuple (dim1, dim2, ...) to compare the array's shape against.
    """
    print(f"--- Audit Report for NumPy Array: {arr_name} ---")
    print(f"Shape: {arr.shape}")
    print(f"Data Type: {arr.dtype}")
    print("First 5 Elements:")
    print(arr[:5])
    print(f"Total NaN Values: {np.isnan(arr).sum()}")

    if expected_shape is not None:
        match_status = "matches" if arr.shape == expected_shape else "does NOT match"
        print(f"Expected Shape: {expected_shape}, Actual Shape {arr.shape}. Shapes {match_status}.")
    print(f"--- End Audit Report for NumPy Array: {arr_name} ---\n")

print("Audit helper functions (audit_df, audit_array) defined.")
````
